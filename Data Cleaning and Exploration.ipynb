# Data Cleaning

```python
for col in data.columns:
    print(data[col].value_counts())
```

```python
# List only columns that have null values
null_columns = data.columns[data.isna().any()]
null_columns
```

```python
print('Percentage of Missing Values in Columns')
for col in null_columns:
    print(col + ':')
    print(data[col].count() /len(data[col]))
    print()
```
